```python
# Bibliotecas
from functions.process_data import read_txt, process_final
```

```python
# INPUTS

file_folder = 'data'

Temp_inicial = 50
Temp_final = 875
Temp_interval = 50

Smoth_DTG = 200
Smoth_TGA = 200
```

```python
# 1. Leitura dos arquivos e geração do dicionário
dict1 = read_txt(data_dir=file_folder,
                 temperatura_inicial=Temp_inicial,
                 temperatura_final=Temp_final,
                 smoth_DTG=Smoth_DTG,
                 smoth_TGA=Smoth_TGA)

print(dict1)
```
